### 2.1. TF-IDF
TF-IDF (*Term Frequency-Inverse Document Frequency*) é uma técnica amplamente utilizada na mineração de texto transformar texto em uma representação numérica. No contexto da análise de sentimentos, o TF-IDF é usado para extrair características relevantes dos textos. O método combina duas métricas: *Term Frequency* (TF) e *Inverse Document Frequency* (IDF). A TF mede a frequência com que um termo aparece em um documento, refletindo a importância do termo no contexto daquele documento específico. Já a IDF avalia a raridade do termo em todo o conjunto de documentos, reduzindo a importância de termos comuns que aparecem em muitos documentos. A fórmula do TF-IDF é o produto de TF e IDF, atribuindo maior peso aos termos que são frequentes em um documento, mas raros em outros, tornando-os mais relevantes para a classificação (Sathya; Mythili, 2023).

Para aplicar TF-IDF na análise de sentimentos, cada texto é convertido em um vetor de valores TF-IDF e esses vetores são então usados como entrada para algoritmos de aprendizado de máquina.

### 2.2. Técnicas utilizadas
Até o momento, as técnicas utilizadas foram Naive Bayes, regressão logística e árvores de decisão. O método Naive Bayes calcula a probabilidade de um texto pertencer a uma categoria de sentimento com base nas probabilidades condicionais das palavras no texto, assumindo que a presença de uma palavra em particular é independente da presença de qualquer outra palavra. Estas probabilidades são aprendidas a partir de um conjunto de dados de treinamento rotulado. Para classificar um novo texto, o modelo multiplica as probabilidades condicionais das palavras presentes no texto pela probabilidade a priori de cada classe de sentimento e escolhe a classe com a maior probabilidade resultante (Villarroal, 2020).

A regressão logística é um método estatístico utilizado para modelar a probabilidade de uma variável dependente binária com base em uma ou mais variáveis independentes. No caso da análise de sentimentos, a variável dependente pode ser a categoria de sentimento. O modelo calcula a combinação linear ponderada das variáveis independentes e aplica a função sigmoide (ou softmax para múltiplas classes) para converter essa combinação em uma probabilidade. Durante o treinamento, os pesos das variáveis são ajustados para minimizar a função de custo, geralmente a entropia cruzada, utilizando algoritmos de otimização como gradiente descendente. O resultado é um modelo que pode prever a probabilidade de uma nova observação pertencer a cada classe de sentimento, facilitando a classificação automática de novos textos (Kausar; Fageeri e Soosaimanickam, 2023).

Outro método estudado foi o de árvores de decisão, o processo começa com a divisão do conjunto de dados em subconjuntos baseados no valor de uma característica, escolhida de forma a maximizar a separação entre as classes. Este procedimento é repetido recursivamente, criando uma estrutura de árvore onde cada nó interno representa uma característica de entrada e cada ramo representa uma regra de decisão, enquanto cada folha representa um resultado da classificação. A divisão é geralmente feita utilizando critérios como a impureza de Gini ou o ganho de informação. Uma vez treinada, a árvore de decisão pode ser usada para classificar novos textos ao seguir o caminho das decisões desde a raiz até uma folha, baseando-se nas características do texto (Kausar; Fageeri e Soosaimanickam, 2023).

O SVM textos em vetores de características e busca um hiperplano que separa os dados em diferentes classes, maximizando a margem de separação entre as classes. Pontos de dados mais próximos do hiperplano, chamados de vetores de suporte, influenciam a definição do hiperplano. Para dados não linearmente separáveis, o SVM utiliza funções de kernel para mapear os dados em um espaço de alta dimensão onde a separação linear é possível (Nurdin; Alexandri; Sumadinata; Arifianti, 2023).

## 3. Métricas de avaliação

Algumas métricas de avaliação que podem ser utilizadas para controle e comprovação de eficiência dos modelos gerados são acuracidade, *precision* e *recall*, cada uma dessas podendo ser obtidas a partir da matriz de confusão. Uma matriz de confusão da biblioteca scikit-learn possui os valores reais nas linhas e os valores previstos pelo modelo nas colunas, portanto, os valores na diagonal da matriz são os valores acertados e os outros são classificados como falsos positivos ou falsos negativos (Villarroal, 2020).

O valor de acuracidade é o resultado da quantidade de acertos calculado com a divisão do total da diagonal da matriz pela quantidade total de linhas utilizadas no teste. Já os valores de *precision* e *recall* possuem as seguintes fórmulas:

Precision: Verdadeiro Positivo / (Verdadeiro Positivo + Falso Positivo)

Recall: Verdadeiro Positivo / (Verdadeiro Positivo + Falso Negativo)

# **Implementação**

Código da solução

Import de bibliotecas necessárias

```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix ##
from sklearn.metrics import accuracy_score
import pickle
import re
```

Faz a leitura do dataset de treinamento e renomeia as colunas

```python
# id arquivo google drive
file_id = '12Ef-EJkUykrHbLU_Ty8yhwhbWCP0IlAM'

url = f'https://drive.google.com/uc?export=download&id={file_id}'
df = pd.read_csv(url)

df = df.rename(columns={df.columns[0]: 'num', df.columns[1]: 'Date', df.columns[2]: 'Title',df.columns[3]: 'Snippet',df.columns[4]: 'Sentiment'})
```

Instala a biblioteca emoji para futuro import

```python
pip install emoji
```

Cria funções de pré-processamento como: substituição de URLs para termo genérico, anonimização de usuários e substituição de emojis por termos.

```python
import emoji
def replace_url(text):
  if isinstance(text, str):
    return re.sub(r'https://t.co/\w+', 'url', text)
  else:
    return text

def replace_username(text):
  if isinstance(text, str):
    return re.sub(r'@\w+', '@user1', text)
  else:
    return text

def replace_emoji(text):
  if isinstance(text, str):
    return emoji.demojize(text)

  else:
    return text
```

Aplica funções de pré processamento

```python
df['Snippet'] = df['Snippet'].apply(replace_username)
df['Snippet'] = df['Snippet'].apply(replace_url)
df['Snippet'] = df['Snippet'].apply(replace_emoji)
```

Verifica se existem valores de rótulo diferentes de positive, neutral e negative.

```python
# lista de tipos de página permitidos
sentimentPermit = ['positive', 'negative', 'neutral']
# verifica se existem valores na coluna 'Page Type' que não estão na lista de permitidos
if not df['Sentiment'].isin(sentimentPermit).all():
    df = df[df['Sentiment'].isin(sentimentPermit)]
    df = df.reset_index(drop=True)
```

Apresenta quantidade de linhas para cada sentimento

```python
df.Sentiment.value_counts()
```

Define as variáveis X e y

```python
train_data = df
train_data = train_data.dropna(subset=['Snippet', 'Sentiment'])

X = train_data['Snippet']
y = train_data['Sentiment']
```

Cria pipeline com vetorização TF-IDF para transformação e classificador Naive Bayes

```python
#pipeline com vetorização TF-IDF e classificador de naive bayes
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB()) #naive bayes
])
```

Separa variáveis de teste e treino e ajusta ao pipeline

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)

pipeline.fit(X_train, y_train)
```

Cálculo e impressão dos resultados

```python
import pickle
with open('naive_bayes_model.pkl', 'wb') as file:
  pickle.dump(pipeline, file)

y_true = y_test
y_pred = pipeline.predict(X_test)

#relatório e matriz de confusao
print("NAIVE BAYES")
print("\n\n\nconfusion matrix")
print(confusion_matrix(y_true, y_pred))
print("\nclassification report")
print(classification_report(y_true, y_pred))
```

Cria pipeline com vetorização TF-IDF para transformação e classificador de regressão logística

```python
#pipeline com vetorização TF-IDF e classificador de arvore de decisão
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', LogisticRegression()) # arvore de decisao
])
```

Separa variáveis de teste e treino e ajusta ao pipeline

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)

pipeline.fit(X_train, y_train)
```

Cálculo e impressão dos resultados

```python
with open('logistic_regression_model.pkl', 'wb') as file:
  pickle.dump(pipeline, file)

y_true = y_test
y_pred = pipeline.predict(X_test)

'''
y_true = y_train
y_pred = pipeline.predict(X_train)

'''
#relatório e matriz de confusao
print("REGRESSÃO LOGÍSTICA") # 76.2
print("\n\nconfusion matrix")
print(confusion_matrix(y_true, y_pred))
print("\nclassification report")
print(classification_report(y_true, y_pred))
```

Cria pipeline com vetorização TF-IDF para transformação e classificador de árvore de decisão

```python
#pipeline com vetorização TF-IDF e classificador de arvore de decisão
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', DecisionTreeClassifier()) # arvore de decisao
])
```

Separa variáveis de teste e treino e ajusta ao pipeline

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)

pipeline.fit(X_train, y_train)
```

Cálculo e impressão dos resultados

```python
with open('decision_tree_model.pkl', 'wb') as file:
  pickle.dump(pipeline, file)

y_true = y_test
y_pred = pipeline.predict(X_test)

'''
y_true = y_train
y_pred = pipeline.predict(X_train)

'''
#relatório e matriz de confusao
print("ÁRVORE DE DECISÃO")
print("\n\nconfusion matrix")
print(confusion_matrix(y_true, y_pred))
print("\nclassification report")
print(classification_report(y_true, y_pred))
```

Os próximos blocos de código apresentam a mesma rotina, mas com diferentes algoritmos como: Random Forest, SVM e Catboost

```python
from sklearn.ensemble import RandomForestClassifier
#pipeline com vetorização TF-IDF e classificador de random forest
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', RandomForestClassifier()) # random forest
])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)

pipeline.fit(X_train, y_train)

with open('decision_tree_model.pkl', 'wb') as file:
  pickle.dump(pipeline, file)

y_true = y_test
y_pred = pipeline.predict(X_test)

'''
y_true = y_train
y_pred = pipeline.predict(X_train)
'''
#relatório e matriz de confusao
print("Floresta Randomica")
print("\n\nconfusion matrix")
print(confusion_matrix(y_true, y_pred))
print("\nclassification report")
print(classification_report(y_true, y_pred))
```

```python
from sklearn.svm import SVC
#pipeline com vetorização TF-IDF e classificador de SVM
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', SVC()) # SVM
])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)

pipeline.fit(X_train, y_train)

with open('decision_tree_model.pkl', 'wb') as file:
  pickle.dump(pipeline, file)

y_true = y_test
y_pred = pipeline.predict(X_test)

'''
y_true = y_train
y_pred = pipeline.predict(X_train)
'''
#relatório e matriz de confusao
print("SVM") # 76.4
print("\n\nconfusion matrix")
print(confusion_matrix(y_true, y_pred))
print("\nclassification report")
print(classification_report(y_true, y_pred))
```
